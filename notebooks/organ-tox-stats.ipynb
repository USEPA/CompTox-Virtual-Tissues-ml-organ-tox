## Compare CHR and SUB

```python
# Compare CHR and SUB chronic effects 
sub_chr = [i for i in Outcomes if re.search('^CHR|^SUB',i)]

X = pd.DataFrame(list(DB.ml_summary_v1.find({'pred':{'$in':sub_chr},
                                             'pt':'cvt','n_obs':100,                                             
                                             'n_ds':{'$lt':25}},
                                            dict(_id=0,pt=0,n_pos=0,n_neg=0,dt_out=0))))

Res = None
for (pred),P_i in X.groupby(['pred']):
    for (dt), P_ij in P_i.groupby(['dt_in']):
        for (lr),P_ijk in P_ij.groupby(['lr']):
            Res = pd.concat((Res,P_ijk.sort('f1_mn',ascending=False).reset_index(drop=True).head(1)))
Perf1 = Res.reset_index(drop=True)


Perf1['res2']=Perf1.apply(lambda x: "%s:%d/%s/%s:%d" % (x.pred,x.n_obs,x.lr,x.dt_in,x.n_ds),axis=1)
Perf1['res2_f1']=Perf1.apply(lambda x: "%3.2f±%3.2f" % (x.f1_mn,x.f1_sd),axis=1)
Perf1['res2_sn']=Perf1.apply(lambda x: "%3.2f±%3.2f" % (x.sens_mn,x.sens_sd),axis=1)
Perf1['res2_sp']=Perf1.apply(lambda x: "%3.2f±%3.2f" % (x.spec_mn,x.spec_sd),axis=1)

P1 = None
for p in Perf1.pred.unique():
    P1 = pd.concat((P1,Perf1.query("pred=='%s'" % p).sort('f1_mn',ascending=False).head(1)))

Perf1['organ']=Perf1.pred.apply(lambda x:x.split(":")[1])
Perf1['study']=Perf1.pred.apply(lambda x:x.split(":")[0])


P2 = pd.pivot_table(Perf1,index=['organ','dt_in','n_ds','lr'],
               columns=['study'],
               values='f1_mn',
               aggfunc=np.mean)

```

Calculate the percentage changes in mean F1 scores between subchronic and chronic studies. Exclude the NB algorithm  and chm descriptors because they produce low F1 scores

```python
P2 = pd.pivot_table(Perf1,index=['organ','dt_in','n_ds','lr'],
               columns=['study'],
               values='f1_mn',
               aggfunc=lambda i: np.unique(i)[0])
P2['d']=P2.apply(lambda x: 100*(x[0]-x[1])/x[0],axis=1)
P2['d1']=P2.d.abs()
R = None
P21 = P2.query("lr!='NB' and dt_in!='chm'").reset_index()
for (o),X in P21.groupby(by='organ'):
    R = pd.concat((R,X.sort('d1',ascending=False).reset_index(drop=True).head(1)))
R['d1']=R.d1.round()
Table_sub_vs_chr = R.sort('d').dropna()
```

# Performance improvement due to more data

```python
X = pd.DataFrame(list(DB.ml_summary_v1.find({'pred':{'$in':Outcomes},
                                             'pt':'cvt','n_obs':{'$gte':100},
                                             '$and':[{'n_ds':{'$gt':2}},{'n_ds':{'$lt':25}}]},
                                                dict(_id=0,pt=0,n_pos=0,n_neg=0,dt_out=0))))


F1=pd.pivot_table(X,columns='n_obs',index=['pred','lr'],values='f1_mn',aggfunc=np.mean)
Sn=pd.pivot_table(X,columns='n_obs',index=['pred','lr'],values='sens_mn',aggfunc=np.mean)
Sp=pd.pivot_table(X,columns='n_obs',index=['pred','lr'],values='spec_mn',aggfunc=np.mean)
N1=F1.max(axis=1)==100
DF1=F1.apply(lambda x: (x.max()-x.min())/x.max(),axis=1)
DSn=Sn.apply(lambda x: (x.max()-x.min())/x.max(),axis=1)
DSp=Sp.apply(lambda x: (x.max()-x.min())/x.max(),axis=1)


Obs=F1.notnull().apply(lambda x: F1.columns[np.where(x)].max(),axis=1)

Y2 = pd.DataFrame(dict(DF1=DF1,DSn=DSn,DSp=DSp,n_obs=Obs))
```

```python
# Now pick the best performing algorithms and calculate the improvement with n_obs
Res=None
for (pred,Y3) in Y2.query("lr != 'NB'").groupby(level=[0]):
    Res = pd.concat((Res,Y3.sort(['DF1'],ascending=False).head(1)))
Res.sort('DF1',ascending=False,inplace=True)
```
